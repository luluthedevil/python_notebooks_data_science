##Estruturas de decisão

Durante o desenvolvimento, pode existir a necessidade de verificar valores e fazer escolhas dependendo dos valores

Para `if`s encadeados, usamos o `elif`

##Estruturas de repetição

Um oitro caso de controle de fluxo, são os loops ou estruturas de repetição.

São usados para que um mesmo bloco de código seja executado para mais de um elemento ou valor

###For

Usado para percorrer uma sequência:
- Um intervalo de números
- Elementos de uma lista

O iterator começa no início da sequência e a cada iteração, é alterado para o próximo valor da sequência

```
for interator in sequencia:
  #comandos a serem executados

```


Podemos usar a função `range(start, end, step)` para nos auxiliar

```
for i in range(1, 11):
  if(i % 2 == 0):
    print("É par")
  else:
    print("É ímpar")
```

No range, o valor `end` não é incluído na lista

Caso o valor `start` não seja especificado, seu valor default é 0

O valor especificado no `step` decide o número de pulos que o iterador dá em cada iteração. Caso não seja especificado, seu valor é 1

````python
seq = [1,2,3,4,5]
````

````python
for item in seq:
    print(item)
````

````python
for item in seq:
    print('Yep')
````

````python
for jelly in seq:
    print(jelly+jelly)
````

#Exceções

##Tratamento de exceções

Quando um erro acontece durante a execução de um programa, uma exceção é lançada

Esse objeto de exceção informa sobre o erro o que aconteceu

Não devem ser tratadas como inimigas, mas sim, tentar proteger o código onde as exceções podem acontecer

Problemas podem acontecer durante a execução dos programas

Podemos proteger nossos códiogos através dos blocos `try-except`

Lista com exceções mais comuns

https://www.w3schools.com/python/python_ref_exceptions.asp

###Try-except

````python
try:
  print(x)
except:
  print("An exception occurred")
````

###try-excep-else

````python
try:
  print('Hello')
except:
  print('Something went wrong')
else:
  print('Nothing went wrong')
````

###try-except-finally

````python
try:
  print(x)
except:
  print("Something went wrong")
finally:
  print('Try-except is finished')
````

###Raise

````python
x = -1

if (x < 0):
  raise Exception("Sorry, no numbers below zero")
````
