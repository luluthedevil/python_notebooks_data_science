##Repeating structures

Another case of flow control are loops or repeating structures. They are used for the same code block to be executed for more than one element or value

###For

Used to step through a sequence:
- A range of numbers
- Elements of a list

The iterator starts at the beginning of the sequence and with each iteration, it changes to the next value in the sequence.

```
for in-sequence interactor:
   # code to run

```

We can use the `range(start, end, step)` function to help us

```
for i in range(1, 11):
   if(i % 2 == 0):
     print("It's even")
   else:
     print("It is odd")
```

In range, the `end` value is not included in the list

If the `start` value is not specified, its default value is 0

The value specified in `step` decides the number of jumps the iterator takes in each iteration. If not specified, its value is 1

````python
seq = [1,2,3,4,5]
````

````python
for item in seq:
    print(item)
````

````python
for item in seq:
    print('Yep')
````

````python
for jelly in seq:
    print(jelly+jelly)
````

#Exceptions

##Exception Handling

When an error occurs during the execution of a program, an exception is thrown.

This exception object tells about the error what happened

They should not be treated as the enemy, but instead, try to protect the code where exceptions can happen.

Problems can happen while running the programs

We can protect our codes through `try-except` blocks

List of most common exceptions

https://www.w3schools.com/python/python_ref_exceptions.asp

###Try-except

````python
try:
  print(x)
except:
  print("An exception occurred")
````

###try-excep-else

````python
try:
  print('Hello')
except:
  print('Something went wrong')
else:
  print('Nothing went wrong')
````

###try-except-finally

````python
try:
  print(x)
except:
  print("Something went wrong")
finally:
  print('Try-except is finished')
````

###Raise

````python
x = -1

if (x < 0):
  raise Exception("Sorry, no numbers below zero")
````
